##  Корректно ли применение линейной регрессии в рамках данной задачи?

Для ответа на этот вопрос необходимо проверить несколько условий:  
1. **Линейность зависимости**

2. **Гомоскедастичность**
	$$De = \sigma^{2}I$$
    
3. **Нормальность ошибок**
	$$e \sim N_{n}(0,\sigma^{2}I)$$

4. **Отсутствие мультиколлинеарности**
	$$rank(X) = m + 1 $$

### Линейность 

Для каждого предиктора необходимо проверить линейность зависимости целевой переменной от него,  
если она нелинейна, преобразовать признак в линейный вид или применять другую модель, учитывающую найденную нелинейность  
(в частных случаях возможно применение полиномиальной регрессии).

#### Методы обнаружения нелинейности 

На практике применяются два основных метода

- **графический**   
  - визуализация зависимости между предиктором и целевой переменной,  
    при слишком большом количестве точек обычно они группируются по бакетам
- **через невязки**  
  - строится простая регрессия по одному предиктору, предсказывающая таргет,  
    затем визуализируются отклонения этой модели (по оси x - оценка целевой переменной,  
    по оси y - отклонение). Критерием линейности считаем симметричность отностельно прямой y = 0.  
    **Замечание**: ось x можно отсчитывать и относительно значений самого признака, вид графика от этого не изменится
    
В качестве метрики оценки линейности признака можно использовать r_2 простой линейной регрессии  
на выбранном признаке относительно целевой переменной. Данную метрику можно интерпретировать,  
как долю зависимости между предиктором и таргетом,можно обнаружить линейной моделью.


#### Способы решения проблемы с нелинейностью

- непрерывное монотонное преобразование, линеаризующее признак (в частных случаях бывают и другие типы функций,  
например, парабола)
- разбиение переменной на линейные фрагменты

```python
from sklearn.linear_model import LinearRegression

def simple_reg(predictor, target):
    '''вспомогательная функция, отрисовка прямой для простой регресии'''
    check = LinearRegression()
    check.fit(predictor.reshape(-1,1),target)
    return hv.Curve((np.array([i for i in np.linspace(min(predictor),max(predictor),100)]),
                     check.coef_*np.array([i for i in np.linspace(min(predictor),
                                                                  max(predictor),100)]) + check.intercept_))
```

##### Разбиение на бакеты

```python
#пример разбиения на бакеты
check = np.ceil(df.monthly_income_amt.rank(pct = True) * 100)
pd.Categorical(check, categories=np.sort(check.unique()),
                    ordered=True)
agg =  df.monthly_income_amt.groupby(check).agg(['min', 'max'])\
         .assign(nums = check.value_counts())
agg.head()
```

```python
def make_bucket(df,feature, n =  100):
    '''функция, бьющая на бакеты(по умолчанию 100 точек)'''
    return df.assign(bucket = np.ceil(df[feature].rank(pct = True) * n))
```

#####  Метод проверки линейности через простую регрессию 

**Невязка** - разность между истинным значением зависимой переменной и ее предсказанием, синоним отклонения

```python
def get_residuals(Y,Pred):
    '''Подсчет невязки Y - истинное значение таргета, Pred - предсказанное'''
    return Y - Pred
```

**Идея**: 
- Для проверямого на линейность признака построить простую линейную регрессию предсказывающую таргет
- подсчитать $\hat{y}$ и невязку
- визуализировать наблюдения, считая по оси абсцисс $\hat{y}$ или значения признака(эквивалентно),  
  по оси y - отклонение
- В качестве критерия линейности берем симметричность относительно прямой y = 0

```python
def check_linearity(predictor,target,f_name,diff):
    simple_model = LinearRegression()
    simple_model.fit(predictor.reshape(-1,1),target)
    predicts = simple_model.predict(predictor.reshape(-1,1))
    return (hv.Scatter(zip(predicts,get_residuals(target,predicts)),\
              kdims = ['Y_estimate'],vdims=['Residual'], 
                       label = f'Проверка линейности признака {f_name}').opts(plot=dict(width=1200,height=900))  
          *hv.Curve(((diff[0] + 0.0001,diff[1] - 0.0001),(0,0))).opts(plot=dict(width=1200,height=900)))
```

### Проверим гомоскедастичность

Необходимо проверить, что признаки гомоскедастичны  
То есть дисперсия ошибки константна на всем предсказании и формально
$$De = \sigma^{2}I$$
Нарушение данного условия приводит к неэффективности оценок регрессии и несоответствию оценок МНК и ОМП

**Способы проверки**: 
- построить график зависимости невязки от предсказания (по необходимости с разбиением на бакеты)
- зависимость разброса от х будет свидетельствовать о гетероскедастичности  
- статистические тесты (к сожалению, универсального нет - каждый позволяет найти гетероскедастичность только определенного вида)

**Способы решения проблемы**
- Преобразование целевой переменной, приводящее к равномерности разброса
- В случае диагональности матрицы ковариаций возможно применение обощенного [МНК](https://studme.org/72649/ekonomika/obobschennyy_metod_naimenshih_kvadratov) (реализовано в [statsmodels](https://www.statsmodels.org/dev/glm.html))
  $$\hat{a} = (X^{T}\Omega^{-1}X)^{-1}X^{T}\Omega^{-1}y, \;где\; \Omega^{-1} - матрица\; ковариаций\; остатков$$

#####  Графический тест

Рисуем тот же график, что и в пункте про проверку линейности через простую регрессию,  
Гетероскедастичность будет наблюдаться, если при изменении y будет меняться величина разброса

```python
def check_homoscedacity(predictor,target,f_name):
    simple_model = LinearRegression()
    simple_model.fit(predictor.reshape(-1,1),target)
    predicts = simple_model.predict(predictor.reshape(-1,1))
    return (hv.Scatter(zip(predicts,get_residuals(target,predicts)),\
              kdims = ['Y_estimate'],vdims=['Residual'], 
                       label = f'Проверка гомоскедастичности признака {f_name}')
            .opts(plot=dict(width=1200,height=900))  
)
```

## Проверим мультиколлинеарность

```python
features = ['accnt_cnt','credit_amt_max','low_age','high_age']
```

Искомые веса находятся по формуле$$\hat{a} = (X^{T}X)^{-1}X^{T}y$$
Необходимо, чтобы матрица X была обратима, то есть $$rank(X) = m + 1$$
Иначе уравнение не будет иметь решения.
Мультиколлинеарность - проблема близости матрицы X к какой-либо матрице неполного ранга  
(по сути сильной зависимости признаков между собой), в данной ситуации решения будут неустойчивыми

#### Первое приближение

Бывает полезно первым шагом посмотреть на матрицу ковариаций признаков

```python
from scipy.stats.stats import pearsonr   
import seaborn as sns

pcorr = pd.DataFrame 
pcorr = []
for f1 in features:
    for f2 in features:
        pcorr.append((f1,f2,pearsonr(df_1[f1],df_1[f2])))
        
cor = df_1.loc[:,features].corr(method='pearson')
cm=plt.cm.viridis
sns.heatmap(cor,cmap=cm,linewidths=0.1,linecolor='white',annot=True)
plt.show()
```

Считаем признаки сильно скоррелироваными при значениях корреляции больших 0.4 (эмпирическая оценка)

#### VIF
Более универсальный критерий - variance_inflation_factor
Позволяет учесть корреляцию признака со всеми остальными в совокупности.  
Для каждого предиктора строится модель из остальных предикторов, предсказывающая его  
(ясно, что признак нескоррелирован при низком R_2)  
$$VIF_{i} = 1/TOL_{i}$$

$$TOL_{i} = 1 - R^{2}_{i}$$


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

pd.DataFrame(data = {'feature':[i for i in features],
                     'VIF':[round(variance_inflation_factor(df[features].values,i),2) for i in range(4)]}).set_index('feature')
```

Обычно считают, что значения больше 10 сигнализируют о мультиколлинеарности, между 5 и 10 с подозрениями на нее.  
В таблице видно, что признаки low_age и high_age сильно скоррелированы, данная корреляция искусственна вонзикает  
из построения.

```python
import statsmodels.formula.api as smf

def forward_selected(data, response):
    """Linear model designed by forward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by forward selection
           evaluated by adjusted R-squared
    """
    remaining = set(data.columns)
    remaining.remove(response)
    selected = []
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = "{} ~ {} + 1".format(response,
                                           ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop()
        if current_score < best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response,
                                   ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model
```
